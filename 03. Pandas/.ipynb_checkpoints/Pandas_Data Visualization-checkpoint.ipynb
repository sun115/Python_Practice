##### 차트 꾸미기

```python
%matplotlib inline
x=list(range(1,11,1))
y=[]
for i in x:
    y.append(i**2)

plt.plot(x,y, 'y--', lw=2)       # 차트꾸미기(선색,선굵기) + 프린트
plt.plot(x,y, 'go')              
plt.xlabel('index')                # x축 라벨
plt.ylabel('value')                # y축 라벨
plt.title('Simple plot')           # 차트 제목
```

##### 레전드, 어노테이션

```python
%matplotlib inline
x=list(range(1,11,1))
y = [i**2 for i in x]            # 여러개의 선을 그릴 수 있다. 
y2 = [i**2.5 for i in x]         # for문 역전시켜서 list에 적용가능.(ONLY 파이썬)
```

```python
# 차트 꾸미기, legend
plt.plot(x,y,"mo--", label="first")
plt.plot(x,y2,"yo--", label="second")
plt.legend(loc=0)
```

```python
plt.plot(x,y,"bo--", label="before")
plt.plot(x,y2,"co--", label="after")
plt.legend(loc=0)

# 화살표 시작지점
plt.annotate('point',xy=(2,3), xytext=(4,3),arrowprops=dict(color="red", width=1))
```

##### 그리드

```python
%matplotlib inline

# 팝업 창 활용하여 차트 시연
x=list(range(1,11,1))
y=[]
for i in x:
    y.append(i**2)
plt.plot(x,y)

# 차트 꾸미기
plt.plot(x,y, 'y', lw=1.5)              #차트 선색 및 크기설정
plt.plot(x,y, 'mo', )                   #차트 내 'o' 표기
plt.grid(True)                          #그리드 표기
plt.xlabel('index')                     # x축 라벨
plt.ylabel('value')                     # y축 라벨
plt.title('Simple plot')                # 차트 제목
```

### 1-2. 데이터 시각화 고급

##### 서브플롯 그리기

```python
#필요 라이브러리 정의
import matplotlib.pyplot as plt
```

```python
# 팝업 창 활용하여 차트 시연
x=list(range(1,11,1))
y = [i**2 for i in x]
y2 = [i**2.2 for i in x]
```

```python
# 서브플롯
fig = plt.figure(figsize=(10,5))                  # 도화지 사이즈 결정

ax1 = fig.add_subplot(211)                        # 2행 1열의 1번째 (위쪽 차트) | 2행1열-위아래, 1행2열-좌우
plt.plot(x,y, 'yo',lw=1.5, label = '1st')
plt.plot(x,y,'y')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('subplot')

ax2 = plt.subplot(212)                             # 2행 1열의 2번째 (아래쪽 차트)
plt.plot(x,y2, 'mo', lw=1.5, label='2nd')
plt.plot(x,y2,'m')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

##### 이중축 활용하기

```python
# 팝업 창 활용하여 차트 시연
x=list(range(1,11,1))
y = [i**2 for i in x]
y2 = [i**2.2 for i in x]

# 이중축
fig = plt.figure(figsize=(10,5))

ax1 = fig.add_subplot(1,1,1)
plt.plot(x,y, 'b--',lw=1.5, label = '1st')
plt.plot(x,y, 'o')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('subplot')

ax2 = ax1.twinx()                                 # x축 같이 쓰자 ! 양옆 두 개의 y 값.
plt.plot(x,y2, 'c', lw=1.5, label='2nd')
plt.plot(x,y2, 'co')
plt.legend(loc=4)
plt.xlabel('index')
plt.ylabel('value')
```

##### 차트 종류 변경

```python
x=list(range(1,11,1))
y = [i**2 for i in x]

plt.figure(figsize=(7,5))

plt.title('Scatter plt')
plt.scatter(x,y,marker='o', edgecolors='c')
```

##### CountPlot - 특정 컬럼 값의 개수 확인

```python
#pip install seasborn
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')      # tips = seaborn 웹상에서 기본으로 제공하는 data.
                                     #https://seaborn.pydata.org/generated/seaborn.load_dataset.html
sns.countplot(x="day", data = tips)  # 요일별 개수(값) 구하기
```

##### DisPlot - 특정 컬럼 값의 분포도 확인

```python
%matplotlib inline

sns.distplot(tips['total_bill'])
```

##### JointPlot - 2차원 실수형 데이터 시각화

```python
sns.jointplot(x= 'total_bill', y = 'tip', data = tips)
```

```python
# 커널 밀도함수

sns.jointplot(x= 'total_bill', y = 'tip', data = tips, kind='kde')
```

##### Corr - 데이터의 상관관계 확인

```python
# 그냥 dataframe에서도 사용가능. (숫자형 column만)
tips.corr()
```

##### Heatmap - 자료에 대한 상관관계 표현

```python
tc = tips.corr()
sns.heatmap(tc,annot=True)

# 색이 연할수록 연관성이 높다
```

```python
flights = sns.load_dataset('flights')
fp = flights.pivot_table(index='month', columns='year', values = 'passengers')

sns.heatmap(fp)

# 최근 연도일수록 승객 수가 증가하고 있음.
# 또한, 여름에 집중됨.
```

##### Implot - 두 데이터의 상관관계를 linear(선형의) 형태로 표현하는 차트

```python
sns.lmplot(x='total_bill', y = 'tip', data = tips )
sns.lmplot(x='total_bill', y = 'tip', data = tips, hue = 'sex', markers = ['o','v'] )
```
