rolling함수에 문제점 발견. 제품이 바뀔 때 이동평균 날개를 리셋해줘야 한다.

따라서 제품별 이동평균 다시 구하기!

```python
groupKey2 = ["REGIONID","PRODUCT"]
```

```python
groupDataSet = mergeData.groupby(groupKey2)
```

```python
# 각 그룹이 하나씩 리턴됨. (식별번호 별 제품명이 하나의 그룹이 됨.)

oneGroup = groupDataSet.get_group(list(groupDataSet.groups)[0])
```

```python
def groupRolling(oneGroup):
    #len(list(groupDataSet.groups))
    #oneGroup = groupDataSet.get_group(list(groupDataSet.groups)[0])
    # oneGroup.head()

    oneGroup = oneGroup.reset_index(drop=True)
    oneGroup["MA2"] = oneGroup["NEW_QTY_x"].rolling(window=5,center=True).mean()
#     oneGroup["MA2"] = oneGroup["NEW_QTY_x"].rolling(window=5,center=True).mean()

    return oneGroup
```

```python
finalResult = mergeData.groupby(groupKey2).apply(groupRolling)
```

```python
finalResult.head(170)
```

```python
finalResult = finalResult.reset_index(drop=True)
```

```python
finalResult=finalResult.rename(columns = {"NEW_QTY_x":"NEW_QTY", "NEW_QTY_y":"QTY_YEAR"})
finalResult.head()
```
