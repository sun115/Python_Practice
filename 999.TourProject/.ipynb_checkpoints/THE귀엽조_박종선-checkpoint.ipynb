### K-Means Clustering

```python
import pandas as pd
# K-Means 라이브러리
from sklearn.cluster import KMeans
# 시각화 라이브러리
import matplotlib.pyplot as plt
```

```python
# Only needed if you want to display your plots inline if using Notebook
# change inline to auto if you have Spyder installed
%matplotlib inline
```

```python
def elbow(X):
    sse = []
    for i in range(1,11):
        km = KMeans(n_clusters=i, init="k-means++", random_state=0)
        km.fit(X)
        sse.append(km.inertia_)
    plt.plot(range(1,11),sse,marker='o')
    plt.xlabel('cluster number')
    plt.ylabel('SSE')
    plt.show()
```

```python
# K-means Clustering
```

```python
### 기본정보 가지고 있기
base = result01[["hid","pid","month","q10_grade"]]
### 특성정보 분리
features = result01[["q6_2"]]
# "q1","q5","q7","q6_1","q6_4"
# 당일/숙박, 일행존재여부, 총지출비용, 여행한광역시, 이동수단
```

```python
# 클러스터 수 정하기
elbow(features)
```

```python
model = KMeans(n_clusters=3, init="k-means++")
model.fit(features)
### ATL (Above The Line) 가시적, 
### BTL (Below The Line) 비가시적, 신문광고 등
```

```python
predict = pd.DataFrame(model.predict(features))
predict.columns=["cluster"]
```

```python
result = pd.concat([base, features, predict], axis = 1)
result
```
