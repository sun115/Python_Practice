# 3. 데이터 정제

### (-) 값을 0 으로 대체함

```python
import numpy as np
```

```python
selloutData["NEW_QTY"] = np.where(selloutData.QTY < 0, 0, selloutData.QTY)
```

```python
selloutData[selloutData.QTY < 0][selloutData.NEW_QTY!=0]
```

### 연도별 주차수가 다른 Year 가 관찰됨 (53 주차 데이터를 제거)

```python
selloutData.max()
```

```python
selloutData["YEAR"] = selloutData.YEARWEEK.str[0:4]
selloutData["WEEK"] = selloutData.YEARWEEK.str[4:6]
```

```python
refinedSelloutData = selloutData[selloutData.WEEK.astype(int) <= 52]
```

```python
len(selloutData) - len(refinedSelloutData)
```

### 데이터 정렬

```python
sortKey = ["REGIONID", "PRODUCT", "YEARWEEK"]
```

```python
sortedData = refinedSelloutData.sort_values(sortKey)
```

### 데이터 집계 (Group By)

```python
groupKey = ["REGIONID", "PRODUCT"]
```

```python
gfunction = ["mean", "std", "size"]
```

```python
groupData = sortedData.groupby(groupKey)["NEW_QTY"].agg(gfunction).reset_index()
groupData
```

```python
groupData.rename(columns={"mean" : "MEAN", "std" : "STD"}, inplace = True)
```

### 데이터 합치기 (Join)

```python
refinedSelloutData.columns
```

```python
groupData.columns
```

```python
joinKey = ["REGIONID", "PRODUCT"]
```

```python
joinData = pd.merge(refinedSelloutData, groupData, on=joinKey, how="left")
joinData
```

```python
reqcolumns = ["REGIONID", "PRODUCT", "YEARWEEK", "YEAR", "WEEK", "NEW_QTY", "MEAN"]
```

```python
middleResult = joinData[reqcolumns]
```

# 4. 계절성 지수 산출

```python
middleResult["SEASONALITY"] = np.where(middleResult.MEAN != 0,
                                       middleResult.NEW_QTY/middleResult.MEAN, 
                                       0)
```

```python
middleResult
```

```python
finalgroupKey = ["REGIONID", "PRODUCT", "WEEK"]
```

```python
finalResult = middleResult.groupby(finalgroupKey)["SEASONALITY"].agg(["mean"]).reset_index()
```

```python
finalResult.rename(columns={"mean" : "SEASONALITY"}, inplace = True)
finalResult.head()
```

```python
oneGroup = groupData2.get_group(list(groupData2.groups)[0])
```

```python
def mafunction(oneGroup, in_window):

    eachData = oneGroup.reset_index(drop=True)
    eachData["MA"] = eachData.NEW_QTY.rolling(window=in_window, center=True, min_periods=1).mean()

    return eachData
```

```python
WINDOW = 5
```

```python
mafunction(oneGroup, WINDOW)
```

```python
groupData2 = sortedData.groupby(groupKey2).apply(mafunction, (WINDOW))
```
