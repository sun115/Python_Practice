```python
from sklearn.preprocessing import LabelEncoder

le_route = LabelEncoder()
le_opportunity = LabelEncoder()
le_Competitor = LabelEncoder()

salesData["Route_enco"] = le_route.fit_transform(salesData["Route To Market"])
salesData["Oppt_enco"] = le_opportunity.fit_transform(salesData["Opportunity Result"])
salesData["Cmpt_enco"] = le_Competitor.fit_transform(salesData["Competitor Type"])
```

```python
groupKey = ["Region","Supplies Group","Supplies Subgroup"]
```

```python
groupData = salesData.groupby(groupKey)
```

```python
onegroup = groupData.get_group( list(groupData.groups)[0] )
```

```python
eachgroup = onegroup.reset_index(drop=True)
```

```python
eachgroup
```

```python
corrdf = eachgroup.corr()
corrStd = 0.2
features = list(corrdf[ (abs(corrdf["Oppt_enco"] ) >= corrStd) & (corrdf["Oppt_enco"] != 1) ].index)
label = ["Oppt_enco"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
trainingData_features,\
testData_features,\
trainingData_label,\
testData_label = train_test_split(eachgroup[features], eachgroup[label],
                                                  test_size = 0.3, random_state = 3)
```

```python
from sklearn import svm
```

```python
### Define Model
model = svm.SVC(random_state=1)
### Learning (Feature & Label)
model.fit(trainingData_features, trainingData_label)
```

```python
## í…ŒìŠ¤íŠ¸ë°ì´í„° Featureë§Œ ê°€ì§€ê³  ì˜ˆì¸¡í•˜ê¸°
predict = model.predict(testData_features)
predictData = pd.DataFrame(predict)
predictData.columns = ["predict"]

predictData
```

```python
testData_features2 = testData_features.reset_index(drop=True)
testData_label2 = testData_label.reset_index(drop=True)
```

```python
finalResult = pd.concat( [testData_features2, testData_label2, predictData], axis=1 )
finalResult
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
# ê²°ê³¼ í…ŒìŠ¤íŠ¸í•˜ê¸° 
ac_score = accuracy_score(finalResult.Oppt_enco, finalResult.predict)

## ê²°ê³¼ ë¦¬í¬íŠ¸í•˜ê¸°
print("Accuracy =", ac_score)
```

```python
cl_report = classification_report(finalResult.Oppt_enco, finalResult.predict)
print("result =\n", cl_report)
```
