```python
import pandas as pd
import numpy as np
```

### 1. 데이터 불러오기 / 타입통합

```python
customerData = pd.read_csv("../dataset2/customer.csv")
```

### 2. 특성 선정 / 데이터 분리

```python
from sklearn.model_selection import train_test_split
```

```python
features = ["deposit", "stock"]
label = ["label"]
```

```python
train_test_split(customerData[features], test_size = 0.3, random_state = 24)
```

```python
trainingData_features,\
testData_features,\
trainingData_label,\
testData_label = train_test_split(customerData[features], customerData[label],
                                                  test_size = 0.3, random_state = 3)
```

```python
from sklearn import svm
```

```python
### Define Model
model = svm.SVC(random_state=1)
### Learning (Feature & Label)
model.fit(trainingData_features, trainingData_label)
```

```python
## 테스트데이터 Feature만 가지고 예측하기
predict = model.predict(testData_features)
predictData = pd.DataFrame(predict)
predictData.columns = ["predict"]

predictData
```

```python
testData_features2 = testData_features.reset_index(drop=True)
testData_label2 = testData_label.reset_index(drop=True)
```

```python
finalResult = pd.concat( [testData_features2, testData_label2, predictData], axis=1 )
finalResult
```

```python
finalResult.to_csv("../dataset2/finalResult.csv")
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
finalResult.columns
```

```python
# 결과 테스트하기 
ac_score = accuracy_score(finalResult.label, finalResult.predict)

## 결과 리포트하기
print("Accuracy =", ac_score)
```

```python
cl_report = classification_report(finalResult.label, finalResult.predict)
print("result =\n", cl_report)
```
